### Creating Factors

```python
def momentum_factor(df, n):
    df['Momentum_' + str(n)] = df['Close'].pct_change(n)
    df['Long_momentum' + str(n)] = np.where(df['Momentum_' + str(n)] > 0, 1, 0)
    df['Short_momentum' + str(n)] = np.where(df['Momentum_' + str(n)] < 0, -1, 0)
    df = df.dropna()
    return df
```

```python
def moving_average_factor(df, n):
    df['MA_' + str(n)] = df['Close'].rolling(n).mean()
    df['Long_MA' + str(n)] = np.where(df['Close'] > df['MA_' + str(n)], 1, 0)
    df['Short_MA' + str(n)] = np.where(df['Close'] < df['MA_' + str(n)], -1, 0)
    df = df.dropna()
    return df
```

```python
def volatility_factor(df, n):
    df['Volatility_' + str(n)] = df['Close'].pct_change().rolling(n).std()
    df['Long_Volatility' + str(n)] = np.where(df['Volatility_' + str(n)] < 0.1, 1, 0)
    df['Short_Volatility' + str(n)] = np.where(df['Volatility_' + str(n)] > 0.1, -1, 0)
    df = df.dropna()
    return df
```

```python
def volume_factor(df, n):
    df['Volume_' + str(n)] = df['Volume'].rolling(n).mean()
    df['Long_Volume' + str(n)] = np.where(df['Volume'] > df['Volume_' + str(n)], 1, 0)
    df['Short_Volume' + str(n)] = np.where(df['Volume'] < df['Volume_' + str(n)], -1, 0)
    df = df.dropna()
    return df
```

```python
def rsi_factor(df, n):
    df['delta'] = df['Close'].diff()
    df['gain'] = np.where(df['delta'] >= 0, df['delta'], 0)
    df['loss'] = np.where(df['delta'] < 0, abs(df['delta']), 0)
    avg_gain = []
    avg_loss = []
    gain = df['gain'].tolist()
    loss = df['loss'].tolist()
    for i in range(len(df)):
        if i < n:
            avg_gain.append(np.NaN)
            avg_loss.append(np.NaN)
        elif i == n:
            avg_gain.append(df['gain'].rolling(n).mean().tolist()[n])
            avg_loss.append(df['loss'].rolling(n).mean().tolist()[n])
        elif i > n:
            avg_gain.append(((n - 1) * avg_gain[i - 1] + gain[i]) / n)
            avg_loss.append(((n - 1) * avg_loss[i - 1] + loss[i]) / n)
    df['avg_gain'] = np.array(avg_gain)
    df['avg_loss'] = np.array(avg_loss)
    df['RS'] = df['avg_gain'] / df['avg_loss']
    df['RSI'] = 100 - (100 / (1 + df['RS']))
    df['Long_RSI' + str(n)] = np.where(df['RSI'] < 30, 1, 0)
    df['Short_RSI' + str(n)] = np.where(df['RSI'] > 70, -1, 0)
    df = df.dropna()
    return df
```

```python
def macd_factor(df, n_fast, n_slow):
    df['MA_Fast'] = df['Close'].ewm(span=n_fast, min_periods=n_slow).mean()
    df['MA_Slow'] = df['Close'].ewm(span=n_slow, min_periods=n_slow).mean()
    df['MACD'] = df['MA_Fast'] - df['MA_Slow']
    df['Signal'] = df['MACD'].ewm(span=9, min_periods=9).mean()
    df['Long_MACD'] = np.where(df['MACD'] > df['Signal'], 1, 0)
    df['Short_MACD'] = np.where(df['MACD'] < df['Signal'], -1, 0)
    df = df.dropna()
    return df
```

```python
def bollinger_bands_factor(df, n):
    df['MA'] = df['Close'].rolling(n).mean()
    df['BB_up'] = df['MA'] + 2 * df['MA'].rolling(n).std()
    df['BB_dn'] = df['MA'] - 2 * df['MA'].rolling(n).std()
    df['BB_width'] = df['BB_up'] - df['BB_dn']
    df['BB_width_norm'] = df['BB_width'] / df['MA']
    df['Long_BB_width' + str(n)] = np.where(df['BB_width_norm'] < 0.05, 1, 0)
    df['Short_BB_width' + str(n)] = np.where(df['BB_width_norm'] > 0.05, -1, 0)
    df = df.dropna()
    return df
```

```python
def stochastic_oscillator_factor(df, n):
    df['L14'] = df['Low'].rolling(window=n).min()
    df['H14'] = df['High'].rolling(window=n).max()
    df['%K'] = 100 * ((df['Close'] - df['L14']) / (df['H14'] - df['L14']))
    df['%D'] = df['%K'].rolling(window=3).mean()
    df['Long_Stochastic' + str(n)] = np.where(df['%K'] < 20, 1, 0)
    df['Short_Stochastic' + str(n)] = np.where(df['%K'] > 80, -1, 0)
    df = df.dropna()
    return df
```

```python
def william_r_factor(df, n):
    df['Hn'] = df['High'].rolling(n).max()
    df['Ln'] = df['Low'].rolling(n).min()
    df['%R'] = (df['Hn'] - df['Close']) / (df['Hn'] - df['Ln']) * - 100
    df['Long_William' + str(n)] = np.where(df['%R'] < -80, 1, 0)
    df['Short_William' + str(n)] = np.where(df['%R'] > -20, -1, 0)
    df = df.dropna()
    return df
```

### 1) Top 2 and Bottom 2 Coins Strategy

```python
# Create dataframe which is same as signalScores dataframe
signalRank = signalScores.copy()
# Select the top 2 coins with the highest score and bottom 2 coins with the lowest score. Then assign 1 to the top 2 coins and -1 to the bottom 2 coins. Assign 0 to the rest of the coins.
# Create a for loop for every day in the signalRank dataframe
for i in range(len(signalRank)):
    # Fill zero for all values
    signalRank.iloc[i,:] = 0
    # Calculate the top 2 coins
    top2 = signalScores.iloc[i].nlargest(2).index
    # Calculate the bottom 2 coins
    bottom2 = signalScores.iloc[i].nsmallest(2).index
    # Assign 1 to the top 2 coins
    signalRank[top2[0]].iloc[i] = 1
    signalRank[top2[1]].iloc[i] = 1
    # Assign -1 to the bottom 2 coins
    signalRank[bottom2[0]].iloc[i] = -1
    signalRank[bottom2[1]].iloc[i] = -1

# Increase the index of the signalRank date by 1 day
signalRank.index = signalRank.index + pd.DateOffset(days=1)
signalRank
```

```python
# Create a dataframe which is same as signalRank dataframe. Values should be return of the coins
coinReturns = signalRank.copy()
# Fill the coinReturns dataframe with the daily returns of the coins from their own dataframes
coinReturns['BTC'] = btc['daily_range']
coinReturns['ETH'] = eth['daily_range']
coinReturns['SOL'] = solana['daily_range']
coinReturns['XRP'] = xrp['daily_range']
coinReturns['ADA'] = cardano['daily_range']
coinReturns['AVAX'] = avax['daily_range']
coinReturns['DOGE'] = doge['daily_range']
coinReturns['TRX'] = tron['daily_range']
coinReturns['DOT'] = polkadot['daily_range']
coinReturns['LINK'] = chainlink['daily_range']
coinReturns['POLYGON'] = polygon['daily_range']
coinReturns['LTC'] = litecoin['daily_range']
coinReturns
```

```python
# Create a dataframe which is same as signalRank dataframe
coinReturnsRank = signalRank.copy()
# Multiply the signalRank dataframe with the coinReturns dataframe
coinReturnsRank = coinReturnsRank * coinReturns
# Calculate the daily return column using sum(axis=1)
coinReturnsRank['daily_return'] = coinReturnsRank.sum(axis=1)
# Calculate the total capital column. Initial capital is 10,000,000
coinReturnsRank['total_capital'] = 10000000 * (1 + coinReturnsRank['daily_return']).cumprod()
# Delete last row of the dataframe
coinReturnsRank = coinReturnsRank[:-1]
coinReturnsRank
```

```python
# Plot the total capital column
coinReturnsRank['total_capital'].plot()
```

```python
# Calculate total return
total_return = (coinReturnsRank['total_capital'].iloc[-1] - 10000000) / 10000000 * 100
total_return
```

### 3) Top 2 and Bottom 2 Coins Strategy with Weights

```python
# Create dataframe which is same as signalScores dataframe
signalRank_3 = signalScores.copy()
# Select the top 2 coins with the highest score and bottom 2 coins with the lowest score. Then assign 1 to the top 2 coins and -1 to the bottom 2 coins. Assign 0 to the rest of the coins.
# Create a for loop for every day in the signalRank dataframe
for i in range(len(signalRank_3)):
    # Fill zero for all values
    signalRank_3.iloc[i,:] = 0
    # Calculate the top 2 coins
    top2 = signalScores.iloc[i].nlargest(2).index
    # Calculate the bottom 2 coins
    bottom2 = signalScores.iloc[i].nsmallest(2).index
    # Assign 1 to the top 2 coins
    signalRank_3[top2[0]].iloc[i] = signalScores[top2[0]].iloc[i]
    signalRank_3[top2[1]].iloc[i] = signalScores[top2[1]].iloc[i]
    # Assign -1 to the bottom 2 coins
    signalRank_3[bottom2[0]].iloc[i] = signalScores[bottom2[0]].iloc[i]
    signalRank_3[bottom2[1]].iloc[i] = signalScores[bottom2[1]].iloc[i]
# Increase the index of the signalRank_3 date by 1 day
signalRank_3.index = signalRank_3.index + pd.DateOffset(days=1)
signalRank_3
```

```python
# Create dataframe which is same as signalScores dataframe
coinReturnsRank_3 = signalRank_3.copy()
# Multiply the signalRank dataframe with the coinReturns dataframe
coinReturnsRank_3 = coinReturnsRank_3 * coinReturns
# Calculate the daily return column using sum(axis=1)
coinReturnsRank_3['daily_return'] = coinReturnsRank_3.sum(axis=1)
# Calculate the total capital column. Initial capital is 10,000,000
coinReturnsRank_3['total_capital'] = 10000000 * (1 + coinReturnsRank_3['daily_return']).cumprod()
# Delete last row of the dataframe
coinReturnsRank_3 = coinReturnsRank_3[:-1]
coinReturnsRank_3
```
